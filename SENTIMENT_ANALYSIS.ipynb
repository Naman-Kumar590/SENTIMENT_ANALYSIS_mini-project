```python
from transformers import XLMRobertaTokenizer, XLMRobertaForSequenceClassification, XLMRobertaConfig
```

```python
MODEL = "cardiffnlp/twitter-xlm-roberta-base-sentiment"
model = XLMRobertaForSequenceClassification.from_pretrained(MODEL)
```

```python
tokenizer = XLMRobertaTokenizer.from_pretrained(MODEL)
config = XLMRobertaConfig.from_pretrained(MODEL)
```

```python
text = "i had a wonderful day at park today."
```

```python
def predict_sentiment(text: str) -> str:
    processed_text = text
    encoded_input = tokenizer(processed_text, return_tensors='pt')
    output = model(**encoded_input)
    index_of_sentiment = output.logits.argmax().item()
    sentiment = config.id2label[index_of_sentiment]
    return sentiment
```

```python
abc = predict_sentiment(text)
print(abc)
```
